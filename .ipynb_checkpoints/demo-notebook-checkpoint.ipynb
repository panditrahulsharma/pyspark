```python
import psycopg2
conn = psycopg2.connect(database="result", user = "postgres", password = "Meta@123", host = "0.0.0.0", port = "5432")
print("Opened database successfully")
```

```python
"""
CREATE TABLE `orderdetails` (
  `orderNumber` int(11) NOT NULL,
  `productCode` varchar(15) NOT NULL,
  `quantityOrdered` int(11) NOT NULL,
  `priceEach` decimal(10,2) NOT NULL,
  `orderLineNumber` smallint(6) NOT NULL,
  PRIMARY KEY (`orderNumber`,`productCode`),
  KEY `productCode` (`productCode`)
) ENGINE=InnoDB DEFAULT CHARSET=latin1;

"""

cur = conn.cursor()
try:
    cur.execute('''CREATE TABLE orderdetails
          (orderNumber INT PRIMARY KEY NOT NULL,
          productCode  CHAR(50) NOT NULL,
          quantityOrdered  INT  NOT NULL,
          priceEach        float8 NOT NULL,
          orderLineNumber  INT  NOT NULL);''')
    print("Table created successfully")
    conn.commit()
    conn.close()
except:
    cur.execute("rollback")
```
